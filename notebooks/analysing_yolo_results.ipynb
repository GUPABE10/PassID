```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ajustes para visualizaci√≥n
sns.set(style="whitegrid")

# Carga de datos
data = pd.read_csv('../data/training_detector/yolo_results.csv')
```

```python
data.columns = data.columns.str.strip()
```

```python
# Graficando Train DFL Loss per Epoch con Seaborn
plt.figure()
sns.lineplot(data=data, x='epoch', y='train/dfl_loss', marker='o', linestyle='-')
plt.title('Train DFL Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Train DFL Loss')
plt.show()
```

```python
# Graficando Metrics Recall(B) per Epoch con Seaborn
plt.figure()
sns.lineplot(data=data, x='epoch', y='metrics/recall(B)', marker='o', linestyle='-')
plt.title('Metrics Recall(B) per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Metrics Recall(B)')
plt.show()
```

```python
# Graficando mAP50(B) and mAP50-95(B) per Epoch con Seaborn
# Para graficar dos l√≠neas en la misma gr√°fica con Seaborn, primero necesitamos reformatear el DataFrame
melted_data = data.melt(id_vars=['epoch'], value_vars=['metrics/mAP50(B)', 'metrics/mAP50-95(B)'],
                        var_name='Metric', value_name='Value')

plt.figure()
sns.lineplot(data=melted_data, x='epoch', y='Value', hue='Metric', marker='o', style='Metric', dashes=False)
plt.title('mAP50(B) and mAP50-95(B) per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.legend(title='Metric')
plt.show()
```
